```python
import pandas as pd
```

```python
uriage_data = pd.read_csv('2/uriage.csv')
uriage_data.head()
```

```python
kokyaku_data = pd.read_excel('2/kokyaku_daicho.xlsx')
kokyaku_data.head()
```

```python
uriage_data['item_name'].head()
```

```python
uriage_data['item_price'].head()
```

```python
uriage_data['purchase_date'] = pd.to_datetime(uriage_data['purchase_date'])
uriage_data['purchase_month'] = uriage_data['purchase_date'].dt.strftime('%Y%m')
uriage_data['item_name'] = uriage_data['item_name'].str.upper()
uriage_data['item_name'] = uriage_data['item_name'].str.replace(' ', '')
uriage_data['item_name'] = uriage_data['item_name'].str.replace(' ', '')

# uriage_data.pivot_table(index='purchase_month', columns='item_name', aggfunc='size', fill_value=0)
uriage_data.pivot_table(index='purchase_month', columns='item_name', values='item_price', aggfunc='sum', fill_value=0)
```

```python
uriage_data.isnull().any(axis=0)
```

```python
price_is_null = uriage_data['item_price'].isnull()
for item in list(uriage_data.loc[price_is_null, 'item_name'].unique()):
  price = uriage_data.loc[(~price_is_null) & (uriage_data['item_name'] == item), 'item_price'].max()
  uriage_data.loc[(price_is_null) & (uriage_data['item_name'] == item), 'item_price'] = price
uriage_data.head()
```

```python
kokyaku_data['顧客名'].head()
```

```python
kokyaku_data['顧客名'] = kokyaku_data['顧客名'].str.replace(' ', '')
kokyaku_data['顧客名'] = kokyaku_data['顧客名'].str.replace('　', '')

kokyaku_data['顧客名'].head()
```

```python
kokyaku_data['登録日'].astype('str').str.isdigit()
```

```python
is_serial = kokyaku_data['登録日'].astype('str').str.isdigit()

from_serial = pd.to_timedelta(kokyaku_data.loc[is_serial, "登録日"].astype('float'), unit='D') + pd.to_datetime('1900/01/01')
from_string = pd.to_datetime(kokyaku_data.loc[~is_serial, '登録日'])

```

```python
kokyaku_data['登録日'] = pd.concat([from_serial, from_string])
kokyaku_data['登録年月'] = kokyaku_data['登録日'].dt.strftime('%Y%m')

rst = kokyaku_data.groupby('登録年月').count()['顧客名']
print(rst)
```

```python
join_data = pd.merge(uriage_data, kokyaku_data, left_on = 'customer_name', right_on = '顧客名', how = 'left')
join_data = join_data.drop('customer_name', axis=1)
join_data.head()
```

```python
dump_data = join_data[['purchase_date', 'purchase_month', 'item_name', 'item_price', '顧客名', 'かな', '地域', 'メールアドレス', '登録日']]
dump_data.to_csv('2/dump_data.csv', index=False)
```
